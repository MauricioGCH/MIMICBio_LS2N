```python
##Main
import numpy as np
import os

channel = np.load(r"Data\Recordings_per_channel\channel_55.npy")



def read_one_channel(filepath, quantum=0.0050863, skip_s=0, length_s=0):
    Fs = 10000

    filesize = os.path.getsize(filepath)
    total_samples = filesize

    skip_samples = int(skip_s * Fs)
    if length_s == 0:
        length_samples = total_samples - skip_samples
    else:
        length_samples = int(length_s * Fs)

    # Memmap: NO carga todo en memoria
    data = np.load(filepath)
    
    return data[skip_samples:length_samples+skip_samples] * quantum

skip_s = 3000 
length_s = 60
Fs = 10000
signal_float = read_one_channel(filepath = r"Data\Recordings_per_channel\channel_55.npy", skip_s = 3000, length_s = 5)
L = len(signal_float)
time = np.linspace(start=skip_s, stop=skip_s + L/Fs, num=L, endpoint=False)
```

```python
from matplotlib import pyplot as plt
plt.figure(figsize=(15,5))
plt.plot(time, signal_float)
plt.xlabel("Time [s]")
plt.ylabel("Amplitude [µV]")
plt.title(f"Channel #{78}")
plt.plot()
plt.show()
```
