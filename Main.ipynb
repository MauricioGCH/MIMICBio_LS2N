```python
import os

filepath = r"Data\\Hautglucose78.txt"
Fs = 14656
N_channels = 64
data_width_bytes = 2

filesize = os.path.getsize(filepath)
bytes_per_second = N_channels * data_width_bytes * Fs

duration_s = filesize / bytes_per_second

print(f"Taille du fichier : {filesize} octets")
print(f"Durée estimée du signal : {duration_s:.2f} s")
```

```python
import numpy as np
import matplotlib.pyplot as plt


# Lecture
data = np.loadtxt(filepath, delimiter="\t", skiprows=4, encoding="latin1")

# Conversion
time_s = data[:, 0] / 1000.0  # ms → s
amplitude = data[:, 1]

# Affichage
plt.figure(figsize=(15, 5))
plt.plot(time_s, amplitude, color='black')
plt.xlabel("Temps (s)")
plt.ylabel("Amplitude [µV]")
plt.title("Signal exporté ASCII depuis .mcd")
plt.grid(True)
plt.show()
```
